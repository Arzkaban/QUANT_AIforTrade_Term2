```python
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline
 
f, (ax1,ax2) = plt.subplots(figsize = (6,4),nrows=2)
```

```python
import numpy as np
import matplotlib.cm as cm
np.random.seed(20180316)
x = np.random.randn(4, 4)
 
f, (ax1, ax2) = plt.subplots(figsize=(6,6),nrows=2)
 
sns.heatmap(x, annot=True,cmap=cm.coolwarm_r, ax=ax1)
sns.heatmap(x, annot=True, alpha=1.0,center=0.0,annot_kws={"size": 16},linewidths=0.01,
            linecolor='white',cbar=True,cmap=cm.coolwarm_r,ax=ax2)
plt.show()
```
